# Method 1: Using tools decorator

```python
from langchain_core.tools import tool
```

```python
# step 1: create a function

def multiply(a,b):
  """ Multiply two numbers"""
  # highly recommended to add docstring to the function
  return a * b
```

```python
# step 2: add type hints
def multiply(a: int, b: int) -> int:
  """ Multiply two numbers"""
  # highly recommended to add types
  return a * b
```

```python
# step 3: add tool decorator

@tool
def multiply(a: int, b: int) -> int:
  """ Multiply two numbers"""
  # highly recommended to add types
  return a * b
```

```python
result = multiply.invoke({"a": 3, "b": 5})
```

# Method 2: Using structured tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
  a: int = Field(required=True, description="The first number to multiply")
  b: int = Field(required=True, description="The second number to multiply")
```

```python
def multiply(a: int, b: int) -> int:
  return a * b
```

```python
multiply_tool = StructuredTool.from_function(
  func=multiply,
  name="multiply",
  description="Multiply two numbers",
  args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({"a": 4, "b": 6})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
```

# Method 3: Using BaseTool Class

```python
from langchain_core.tools import BaseTool
from typing import Type
from pydantic import BaseModel, Field
```

```python
# arg schema using pydantic
class MultiplyInput(BaseModel):
  a: int = Field(required=True, description="The first number to multiply")
  b: int = Field(required=True, description="The second number to multiply")
```

```python
class MultiplyTool(BaseTool):
  name: str = "multiply"
  description: str = "Multiply two numbers"

  args_schema: Type[BaseModel] = MultiplyInput

  # function name should be _run
  def _run(self, a: int, b: int) -> int:
    return a * b

```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({"a": 4, "b": 6})
```

```python
print(result)
print(multiply_tool.name)
```

# Toolkit

```python
from langchain_core.tools import tool

@tool
def add(a: int, b: int) -> int:
  """Add two numbers"""
  return a + b

@tool
def multiply(a: int, b: int) -> int:
  """Multiply two numbers"""
  return a * b
```

```python
class MathToolkit:

  def get_tools(self):
    return [add, multiply]
```

```python
toolkit = MathToolkit()

tools = toolkit.get_tools()

for tool in tools:
  print(tool.name)
  print(tool.description)
  print(tool.args)
```
