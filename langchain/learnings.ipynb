# Key components:
  1. Models
  2. Prompts
  3. Chains
  4. Indexes
  5. Memory
  6. Agents

---
1. Models:

    Allows to interact with any AI models like openai, claude, gemini, open source hugging face etc
    
2. Prompts:
    
    Input to LLM
    
3. Chains:

    Used for building pipelines
    a. sequential
    b. parallel
    c. conditional

4. Indexes:

    Connects app to external knowledge like pdfs,websites or databases

5. Memory:
    
    llm api calls are stateless

6. Agents

    Used to build AI agents

    Agents are advanced form of chatbots that has reasoning capabilities and tool access

    Eg: A travel AI Agent can check for best tourist place for summer. Find the flights available on a particular day and book the flight tickets




# Temperature
Parameter that controls randomness of model output

| Use Case | Recommended Temperature |
| -------- | ----------------------- |
| Factual answer (math,code,facts) | 0.0 - 0.3 |
| Balanced response (general QA, explanation) | 0.5 - 0.7 |
| Creative writing,story telling, jokes | 0.9 - 1.2 |
| Maximum randomness (wild ideas, brainstorming) | 1.5+ |
